```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from pycaret.regression import *
```

```python
df = pd.read_csv("garments_worker_productivity.csv")
```

```python
df = df.fillna(0)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df['quarter'] = df['quarter'].astype('category')
df['quarter'] = df['quarter'].cat.codes

df['department'] = df['department'].astype('category')
df['department'] = df['department'].cat.codes

df['day'] = df['day'].astype('category')
df['day'] = df['day'].cat.codes

df['date'] = df['date'].astype('category')
df['date'] = df['date'].cat.codes
```

```python
X_train, X_test = train_test_split(df, test_size = .2, random_state = 150)
```

```python
s = setup(data = X_train, test_data = X_test, target = 'actual_productivity',numeric_features = ['team', 'targeted_productivity', 'smv', 'incentive'],  fold_strategy = 'timeseries', fold = 3, session_id = 123)
```

```python
best = compare_models(sort = 'MAE')
```

```python
predictions = predict_model(best, data = X_test)
```
